# Neural Collaborative Filtering

## Роль notebook

Этот notebook должен менять в первую очередь класс модели, а не весь эксперимент.
Его задача: сравнить `Neural CF` с `ALS` на максимально похожей offline-постановке.

## План

- использовать тот же базовый датасет и тот же тип сигнала, что и в `ALS`;
- явно описать negative sampling и ограничения постановки;
- обучить простую `Neural CF` модель;
- посчитать те же top-K метрики;
- сравнить результаты с `ALS` в честных одинаковых условиях.

## Validation checklist

- notebook запускается сверху вниз;
- split и candidate universe согласованы с `ALS`;
- baseline comparison не подменён сравнением разных экспериментов;
- ограничения по сложности и интерпретируемости описаны прямо.
